# STEP #1: IMPORT LIBRARIES/DATASETS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline  
```

```python
#import CIFAR10 dataset
from keras.datasets import cifar10

(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

# STEP #3: DATA PREPARATION

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
number_category = 10
```

```python
y_train
```

```python
#  Using the method to_categorical(), a numpy array (or) a vector which has integers that represent 
# different categories, can be converted into a numpy array (or) a matrix which has binary values and has 
# columns equal to the number of categories in the data.
import tensorflow
y_train = tensorflow.keras.utils.to_categorical(y_train, number_category)
```

```python
y_train
```

```python
y_train.shape
```

```python

y_test = tensorflow.keras.utils.to_categorical(y_test, number_category)
y_test
```

```python
# get values in range of 0 to 1 therefore divide by 255
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
X_train.shape
```

```python
input_shape = X_train.shape[1:]
input_shape
```

# STEP #4: TRAIN THE MODEL

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import adam_v2
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
# Convolutional layer1
cnn_model.add(Conv2D(filters = 32, 
                    kernel_size =(3, 3), 
                    activation = 'relu',
                    input_shape = input_shape))
cnn_model.add(Conv2D(filters = 32,
                    kernel_size = (3, 3),
                    activation = 'relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.3))

# Convolutional layer2
cnn_model.add(Conv2D(filters = 64,
                    kernel_size = (3,3),
                    activation = 'relu'))
cnn_model.add(Conv2D(filters = 64,
                    kernel_size = (3,3),
                    activation = 'relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.2))

cnn_model.add(Flatten())

# Hidden Layer
cnn_model.add(Dense(units = 512, activation = 'relu'))
cnn_model.add(Dense(units = 512, activation = 'relu'))

# Output Layer

## Note: unots for Output should be same as number of categories defined for output i.e. 10
## activation = SOFTMAX because the output should fall(i.e. classify data as 0 or 1) in range of 0 to 1 (mostly probability value)
## whereas RELU is more of Regression values
cnn_model.add(Dense(units = 10, activation = 'softmax'))
```

```python
from tensorflow.keras.optimizers import RMSprop

# loss function

## Note: categorical_crossentropy: Used as a loss function for multi-class classification model 
## where there are two or more output labels. The output label is assigned one-hot category encoding value in 
## form of 0s and 1. The output label, if present in integer form, is converted into categorical encoding using keras.
cnn_model.compile(loss = 'categorical_crossentropy', 
                  optimizer = RMSprop(learning_rate=0.0001),
                  metrics = ['accuracy'])
```

```python
# Train model
history = cnn_model.fit(X_train, y_train, batch_size = 32, epochs = 10, shuffle = True)
```

# STEP #5: EVALUATE THE MODEL

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print("Test Accuracy: {}".format(evaluation[1]))
```

```python
predicted_classes = np.argmax(cnn_model.predict(X_test), axis=-1)
predicted_classes
```

```python
y_test = y_test.argmax(1)
y_test
```

```python
L = 7
W = 7
fig, axes = plt.subplots(L, W, figsize = (12, 12))
axes = axes.ravel()

for i in np.arange(0, L*W):
    axes[i].imshow(X_test[i])
    axes[i].set_title('Prediction = {}\n True = {}'.format(predicted_classes[i], y_test[i]))
    axes[i].axis('off')

plt.subplots_adjust(hspace = 1)    
```

```python
# Summarize results
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize = (10, 10))
sns.heatmap(cm, annot = True)
```

```python
X_train_aug = X_train
X_test_aug = X_test
y_train_aug = y_train
y_test_aug = y_test
```

# STEP #7: IMPROVING THE MODEL WITH DATA AUGMENTATION

- Image Augmentation is the process of artificially increasing the variations of the images in the datasets by flipping, enlarging, rotating the original images. 
- Augmentations also include shifting and changing the brightness of the images.

# STEP 7.1 DATA AUGMENTATION FOR THE CIFAR-10 DATASET

```python
import keras
from keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train.shape
```

```python
n = 10
X_train_sample = X_train[:n]
```

```python
X_train_sample.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator

# dataget_train = ImageDataGenerator(rotation_range = 90)
# dataget_train = ImageDataGenerator(vertical_flip=True)
# dataget_train = ImageDataGenerator(height_shift_range=0.5)
dataget_train = ImageDataGenerator(brightness_range=(1,3))
# shift + Tab for function expansion

dataget_train.fit(X_train_sample)
```

```python
from PIL import Image

fig = plt.figure(figsize = (20,2))
for x_batch in dataget_train.flow(X_train_sample, batch_size = n):
    for i in range(0,n):
        ax = fig.add_subplot(1, n, i+1)
        ax.imshow(Image.fromarray(np.uint8(x_batch[i])))
    fig.suptitle('Augmented images (brightness_range 1 to 3)')
    plt.show()
    break;

plt.subplots_adjust(hspace = 1) 
```

# STEP 7.2 MODEL TRAINING USING AUGEMENTED DATASET

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
                            rotation_range = 90,
                            width_shift_range = 0.1,
                            horizontal_flip = True,
                            vertical_flip = True
                             )
```

```python
datagen.fit(X_train_aug)
```

```python
cnn_model.fit_generator(datagen.flow(X_train_aug, y_train_aug, batch_size = 32), epochs = 2)
```

```python
score = cnn_model.evaluate(X_test_aug, y_test_aug)
print('Test accuracy', score[1])
```
